3. **핵심 개념**
  - **탐색적 데이터 분석(Exploratory Data Analysis, EDA)**: 데이터의 분포, 결측치, 이상치, 변수 간 관계를 시각화하고 분석하여 데이터 특성을 파악하는 과정
  - **결측치 처리(Imputation)**: 누락된 데이터를 평균, 중앙값, 최빈값 등으로 대체하여 데이터 완전성을 확보하는 기법
  - **이상치 탐지 및 제거**: IQR(사분위 범위) 등의 통계적 방법으로 극단값을 식별하고 처리하는 과정
  - **표준화(Standard Scaling)**: 평균 0, 분산 1로 특성을 변환하여 학습 안정성과 수렴 속도를 개선하는 전처리 기법
  - **선형 회귀(Linear Regression)**: 입력 변수와 출력 변수 간의 선형 관계를 모델링하여 연속형 값을 예측하는 지도 학습 알고리즘
  - **회귀 평가 지표**: RMSE(평균 제곱근 오차), MAE(평균 절대 오차), R²(결정계수) 등 모델 성능을 정량화하는 측정 기준

### Concept Check

#### 왜 EDA가 중요한가?

실습에서 우리는 간단한 **탐색적 데이터 분석(Exploratory Data Analysis, EDA)** 를 수행했습니다. EDA는 데이터의 특성을 이해하고 문제를 정의하는 첫 단계입니다. 바로 모델을 학습시키는 것보다, 데이터를 충분히 이해하는 것이 더 나은 모델을 만드는 지름길입니다. 다시 한 번 복습해볼까요?

**EDA에서 확인해야 할 항목들:**
1. **데이터 구조**: 샘플 수, 특성 수, 데이터 타입
2. **기초 통계량**: 평균, 중앙값, 표준편차, 최솟값, 최댓값
3. **결측치(Missing Values)**: 어떤 변수에 얼마나 결측이 있는가?
4. **이상치(Outliers)**: 극단적인 값들이 존재하는가?
5. **분포(Distribution)**: 타겟 변수와 특성들의 분포는 어떤가?
6. **상관관계(Correlation)**: 변수들 간의 선형 관계는 어떠한가?

#### 보스턴 주택 가격 데이터셋

이번 과제에서 사용할 데이터는 1970년대 보스턴 지역의 주택 가격 데이터입니다.

**주요 특성 설명:**
- `CRIM`: 범죄율 (per capita crime rate)
- `ZN`: 25,000 평방피트 이상의 주거지역 비율
- `INDUS`: 비소매 상업지역 비율
- `CHAS`: 찰스강 인접 여부 (1: 인접, 0: 비인접)
- `NOX`: 일산화질소 농도 (공기 오염도)
- `RM`: 평균 방 개수

- `DIS`: 보스턴 고용센터까지의 거리
- `RAD`: 고속도로 접근성 지수
- `TAX`: 재산세율
- `PTRATIO`: 학생-교사 비율
- `B`: 흑인 거주 비율 관련 지수 (윤리적 논란)
- `LSTAT`: 저소득층 비율
- **`MEDV`** (타겟): 주택 가격 중앙값 ($1,000 단위)

### TODO 1: 데이터셋 기초 탐색

데이터의 기본 정보를 파악해봅시다.

**요구사항:**
1. 전체 샘플 수를 `sample_count`에 저장하세요
2. 전체 특성 수(타겟 제외)를 `feature_count`에 저장하세요
3. 타겟 변수 `MEDV`의 평균, 중앙값, 표준편차, 최솟값, 최댓값을 각각 `medv_mean`, `medv_median`, `medv_std`, `medv_min`, `medv_max`에 저장하세요

````python
# 1. 전체 샘플 수를 `sample_count`에 저장하세요
sample_count = len(df)

# 2. 전체 특성 수(타겟 제외)를 `feature_count`에 저장하세요
feature_count = len(df.columns) - 1  # 타겟 제외

# 3. 타겟 변수 `MEDV`의 평균, 중앙값, 표준편차, 최솟값, 최댓값을
# 각각 `medv_mean`, `medv_median`, `medv_std`, `medv_min`, `medv_max`에 저장하세요
medv_mean = df['MEDV'].mean()
medv_median = df['MEDV'].median()
medv_std = df['MEDV'].std()
medv_min = df['MEDV'].min()
medv_max = df['MEDV'].max()

# 검증
EXPECTED_SAMPLES = 506
EXPECTED_FEATURES = 12
assert sample_count == EXPECTED_SAMPLES, f"샘플 수는 {EXPECTED_SAMPLES}이어야 합니다. (got {sample_count})"
assert feature_count == EXPECTED_FEATURES, f"특성 수는 {EXPECTED_FEATURES}이어야 합니다. (got {feature_count})"
assert 20 < medv_mean < 25, f"MEDV 평균이 예상 범위를 벗어났습니다: {medv_mean}"
assert medv_min == 5.0, f"MEDV 최솟값이 5.0이어야 합니다: {medv_min}"
assert medv_max == 50.0, f"MEDV 최댓값이 50.0이어야 합니다: {medv_max}"

print(f"총 샘플 수: {sample_count}")
print(f"총 특성 수: {feature_count}")
print(f"\nMEDV 통계량:")
print(f"  평균: ${medv_mean:.2f}k")
print(f"  중앙값: ${medv_median:.2f}k")
print(f"  표준편차: ${medv_std:.2f}k")
print(f"  범위: ${medv_min:.1f}k ~ ${medv_max:.1f}k")
````

데이터의 품질을 확인하기 위해 결측치와 이상치를 탐색해봅시다.

1. 각 컬럼별 결측치 개수를 계산하여 내림차순으로 정렬한 Series를 `missing_counts`에 저장하세요
2. IQR 방식으로 `MEDV`의 이상치 하한/상한을 계산해 `medv_outlier_lower`, `medv_outlier_upper`에 저장하세요
   - 참고: IQR = Q3 - Q1, 이상치 범위 = [Q1 - 1.5×IQR, Q3 + 1.5×IQR]
3. 이 범위를 벗어난 샘플의 비율(%)을 `medv_outlier_ratio`에 저장하세요

````python
# 1. 각 컬럼별 결측치 개수를 계산하여 내림차순으로 정렬한 Series를 `missing_counts`에 저장하세요
missing_counts = df.isna().sum().sort_values(ascending=False)

# 2. IQR 방식으로 `MEDV`의 이상치 하한/상한을 계산해
# `medv_outlier_lower`, `medv_outlier_upper`에 저장하세요
q1 = df['MEDV'].quantile(0.25)
q3 = df['MEDV'].quantile(0.75)
iqr = q3 - q1
medv_outlier_lower = q1 - 1.5 * iqr
medv_outlier_upper = q3 + 1.5 * iqr

# 3. 이 범위를 벗어난 샘플의 비율(%)을 `medv_outlier_ratio`에 저장하세요
outlier_mask = (df['MEDV'] < medv_outlier_lower) | (df['MEDV'] > medv_outlier_upper)
medv_outlier_ratio = outlier_mask.mean() * 100

# 검증
assert missing_counts.sum() == 0, "보스턴 데이터셋에는 결측치가 없어야 합니다"
assert medv_outlier_lower < medv_outlier_upper, "이상치 하한이 상한보다 작아야 합니다"
assert 0 <= medv_outlier_ratio <= 100, "이상치 비율은 0~100% 범위여야 합니다"

print(f"\n결측치 총 개수: {missing_counts.sum()}")
print(f"\nMEDV 이상치 경계:")
print(f"  하한: ${medv_outlier_lower:.2f}k")
print(f"  상한: ${medv_outlier_upper:.2f}k")
print(f"  이상치 비율: {medv_outlier_ratio:.2f}%")
````

변수들 간의 관계를 파악하여 어떤 특성이 주택 가격에 영향을 미치는지 탐색해봅시다.

**요구사항:**
1. 모든 수치형 변수에 대해 타겟 `MEDV`와의 피어슨 상관계수를 계산하여 `corr_with_medv`에 저장하세요.
2. 자기 자신(MEDV)을 제외하고 절대값 기준 상관계수가 가장 큰 특성명을 `top_corr_feature`에 저장하고, 그 상관계수 값을 `top_corr_value`에 저장하세요
3. 전체 상관행렬에서 $|\text{corr}| >= 0.7$인 특성 쌍(대각선 및 중복 제외)을 `strong_corr_pairs` 리스트에 저장하세요
4. 상관행렬 히트맵을 그려주세요. Arguments: annot=True, fmt=".2f", cmap="coolwarm"

````python
# 1. 모든 수치형 변수에 대해 타겟 `MEDV`와의 피어슨 상관계수를 계산하세요
corr_matrix = df.corr()
corr_with_medv = corr_matrix['MEDV'].drop('MEDV')

# 2. 자기 자신(MEDV)을 제외하고 절대값 기준 상관계수가 가장 큰 특성명을 `top_corr_feature`에 저장하고,
# 그 상관계수 값을 `top_corr_value`에 저장하세요
top_corr_feature = corr_with_medv.abs().idxmax()
top_corr_value = corr_with_medv[top_corr_feature]

# 3. 전체 상관행렬에서 |corr| >= 0.7인 특성 쌍(대각선 및 중복 제외)을
# `strong_corr_pairs` 리스트에 저장하세요
strong_corr_pairs = []
cols = corr_matrix.columns.tolist()
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        if abs(corr_matrix.iloc[i, j]) >= 0.7:
            strong_corr_pairs.append((cols[i], cols[j]))

# 4. 상관행렬 히트맵을 그려주세요. Arguments: annot=True, fmt=".2f", cmap="coolwarm"
fig, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix), k=1)  # 상삼각 마스킹
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            mask=mask, vmin=-1, vmax=1, center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8}, ax=ax)
ax.set_title("Correlation Matrix (Boston Housing)", fontsize=16, pad=20)
plt.tight_layout()
plt.show()

# 검증
assert top_corr_feature in df.columns, "top_corr_feature가 유효한 컬럼명이어야 합니다"
assert -1 <= top_corr_value <= 1, "상관계수는 -1~1 범위여야 합니다"
assert len(strong_corr_pairs) > 0, "강한 상관관계 쌍이 최소 1개 이상 있어야 합니다"

print(f"\nMEDV와 가장 강한 상관관계를 가진 특성: {top_corr_feature}")
print(f"상관계수: {top_corr_value:.4f}")
````

## Step 2: 데이터 전처리 워크플로우

### Concept Check

#### 왜 데이터 전처리가 필요한가?

실제 데이터는 모델이 직접 학습하기에 적합하지 않은 경우가 많습니다. 전처리를 통해:
1. **결측치 처리**: 누락된 데이터를 채워 완전한 데이터셋 구성
2. **이상치 제거**: 극단값이 모델 학습을 방해하지 않도록 제거
3. **스케일링**: 서로 다른 단위의 특성들을 동일한 스케일로 조정
4. **Train/Test 분할**: 일반화 성능을 객관적으로 평가

#### Train/Test Split의 중요성

**데이터 누수(Data Leakage)** 방지를 위해 다음 원칙을 지켜야 합니다:
1. 먼저 train/test를 분할
2. train 데이터로 통계량(평균, 표준편차, IQR 등)을 계산
3. train에서 계산한 통계량을 test에도 동일하게 적용

❌ **잘못된 예**: 전체 데이터로 스케일링 → train/test 분할   
✅ **올바른 예**: train/test 분할 → train으로 스케일러 학습 → train/test 각각 변환

### TODO 4: Train/Test 데이터 분할

모델의 일반화 성능을 평가하기 위해 데이터를 분할합니다.

**요구사항:**
1. 데이터 `df`를 특성과 타겟으로 분리하여 `X`와 `y`에 저장하세요.
2. `train_test_split`을 사용하여 전체 데이터를 다음 조건에 맞춰 분할하세요.
  - 훈련데이터 70%, 검증데이터 30%로 분할하고, 재현성을 위해 `random_state=42`로 설정하세요
  - 타겟 변수 `MEDV`를 기준으로 stratify를 적용하고 싶지만, 연속형 변수라 불가능합니다. 대신 `pd.qcut`으로 10개 구간으로 나눈 후 stratify하세요
  - `X_train`, `X_test`, `y_train`, `y_test`에 각각 저장하세요

````python
# 1. 데이터 `df`를 특성과 타겟으로 분리하여 `X`와 `y`에 저장하세요.
X = df.drop('MEDV', axis=1)
y = df['MEDV']

# 2. `train_test_split`을 사용하여 전체 데이터를 다음 조건에 맞춰 분할하세요.
from sklearn.model_selection import train_test_split
stratify_bins = pd.qcut(y, q=10, labels=False, duplicates='drop')
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=stratify_bins
)

# 검증
expected_train_size = int(len(df) * 0.7)
expected_test_size = len(df) - expected_train_size
assert abs(len(X_train) - expected_train_size) <= 5, f"Train 크기가 예상과 다릅니다: {len(X_train)}"
assert abs(len(X_test) - expected_test_size) <= 5, f"Test 크기가 예상과 다릅니다: {len(X_test)}"
assert len(X_train) == len(y_train), "X_train과 y_train의 크기가 다릅니다"
assert len(X_test) == len(y_test), "X_test와 y_test의 크기가 다릅니다"

print(f"Train 데이터: {len(X_train)}개 샘플")
print(f"Test 데이터: {len(X_test)}개 샘플")
print(f"Train/Test 비율: {len(X_train)/len(df)*100:.1f}% / {len(X_test)/len(df)*100:.1f}%")
````

### TODO 5: IQR 기반 이상치 제거

극단값이 모델 학습에 악영향을 주지 않도록 제거합니다.

**요구사항:**
1. 훈련 데이터의 각 수치형 컬럼에 대해 IQR 기반 이상치 경계를 계산하여 `iqr_bounds` 딕셔너리에 저장하세요
2. 훈련 데이터에서 모든 특성이 정상 범위 내에 있는 데이터를 나타내는 boolean Series `mask_train`을 저장하세요. 그리고 이를 활용해 정상 범위 데이터만 남겨주세요.
3. 검증 데이터도 동일한 방법으로 `mask_test`에 정상 범위 데이터 위치를 표현한 후 정상 범위 데이터만 남겨주세요.
   - **🌟중요🌟**: 훈련 데이터에서 계산한 경계를 검증에도 사용해야 합니다 (데이터 누수 방지)

````python
# 1. 훈련 데이터의 각 수치형 컬럼에 대해 IQR 기반 이상치 경계를 계산하여 `iqr_bounds` 딕셔너리에 저장하세요
iqr_bounds = {}
for col in continuous_cols:
    q1 = X_train[col].quantile(0.25)
    q3 = X_train[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    iqr_bounds[col] = (lower, upper)

# 2. 훈련 데이터에서 모든 특성이 정상 범위 내에 있는 데이터를 나타내는
# boolean Series `mask_train`을 저장하세요. 그리고 이를 활용해 정상 범위 데이터만 남겨주세요.
mask_train = pd.Series(True, index=X_train.index)
for col, (lower, upper) in iqr_bounds.items():
    mask_train &= X_train[col].between(lower, upper)

X_train = X_train[mask_train]
y_train = y_train[mask_train]

# 3. 검증 데이터도 동일한 방법으로 `mask_test`에 정상 범위 데이터 위치를 표현한 후 정상 범위 데이터만 남겨주세요.
mask_test = pd.Series(True, index=X_test.index)
for col, (lower, upper) in iqr_bounds.items():
    mask_test &= X_test[col].between(lower, upper)

X_test = X_test[mask_test]
y_test = y_test[mask_test]

# 검증
assert len(X_train) > 0, "Train 데이터가 모두 제거되었습니다"
assert len(X_test) > 0, "Test 데이터가 모두 제거되었습니다"
assert len(X_train) == len(y_train), "X_train과 y_train의 크기가 다릅니다"

print(f"이상치 제거 후 Train: {len(X_train)}개 샘플")
print(f"이상치 제거 후 Test: {len(X_test)}개 샘플")
````

## Step 3: 모델 학습 및 평가

### Concept Check

#### 선형 회귀 모델의 작동 원리

선형 회귀는 입력 특성들의 선형 조합으로 출력을 예측합니다:

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_n x_n
$$

**학습 과정:**
1. 초기 파라미터($\theta$) 설정 (보통 0 또는 랜덤)
2. 예측값 계산
3. 손실(MSE) 계산
4. 경사 하강법으로 파라미터 업데이트
5. 손실이 수렴할 때까지 2~4 반복

**sklearn의 LinearRegression:**
- 정규방정식(Normal Equation) 또는 SVD를 사용하여 한 번에 최적 해를 구함
- 반복 학습이 필요 없어 빠르지만, 특성 수가 많으면 메모리 부담

#### 모델 평가의 중요성

**Train 데이터 성능만으로는 부족합니다:**
- Train 성능이 좋아도 Test 성능이 나쁘면 과적합
- 실제로 중요한 것은 "본 적 없는 데이터"에 대한 성능

**여러 지표를 함께 봐야 하는 이유:**
- RMSE: 큰 오차에 민감, 타겟과 같은 단위로 해석 용이
- MAE: 이상치에 덜 민감, 평균적인 오차 크기
- R²: 모델의 설명력, 0~1 범위로 직관적

1. 검증 데이터에 대한 예측값을 `y_pred`에 저장하세요
2. RMSE, MAE, R² 지표를 계산하여 각각 `rmse`, `mae`, `r2`에 저장하세요

````python
# 1. 검증 데이터에 대한 예측값을 `y_pred`에 저장하세요
y_pred = model.predict(X_test)

# 2. RMSE, MAE, R² 지표를 계산하여 각각 `rmse`, `mae`, `r2`에 저장하세요
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# 검증
assert rmse > 0, "RMSE는 양수여야 합니다"
assert mae > 0, "MAE는 양수여야 합니다"
assert -1 <= r2 <= 1, "R²는 -1~1 범위여야 합니다"
assert len(y_pred) == len(y_test), "예측값 개수가 테스트 데이터와 다릅니다"

print(f"\n=== 모델 성능 (Test Set) ===")
print(f"RMSE: ${rmse:.3f}k")
print(f"MAE:  ${mae:.3f}k")
print(f"R²:   {r2:.4f}")
````
