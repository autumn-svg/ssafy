```python
class MyClass: #젤 앞에 대문자, 클래스는 설계도
    count = 0 
    #여기에 쓰는 변수는 클래스 변수: MyClass를 통해 생성된 모든 인스턴스가 같은 값을 공유


#인스턴스 메서드: 인스턴스의 상태에 접근해서 작업할 때 사용
    def my_method(self):
        #인스턴스 변수 만들 수 있음: 각각의 인스턴스가 독립된 값을 가지고 있다.
        self.v = 'instance'
        return self


#class 메서드 cls: 클래스의 상태에 접근해서 작업할 때 사용
@classmethod #이걸 안 붙이면 클래스 메서드는 인스턴스 메서드가 됨
def class_method(cls): 
    return cls


#정적 메서드 외부에서 받아오는 값이 없음, 인스턴스 접근x클래스도 안 됨, 두 가지 상태와 관련 없는 작업을 할 때 사용
@staticmethod
def static_method():
    return 'static'

def __str__(self):
    return '나는 마이클래스의 인스턴스다.'


m = MyClass()
m.my_method()
# print(m.count)
# print(m.my_method())
print(MyClass.count, m.v) #count를 출력하고 싶음
#클래스 변수는 클래스 이름을 통해서 접근하자(인스턴스랑 헷갈릴 수 있음)


print(MyClass.class_method())
print(m.my_method())
```

```python
#OOP 2

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def talk(self):
        print(f'반갑습니다. {self.name}입니다.')

    def eat(self):
        print('냠냠쩝쩝')

class Professor(Person): #person 거 물려받겟다
    def __init__(self, name, age, department):
        self.name = name
        self.age = age
        self. department = department #모든 사람이 전공을 가지고 있진 않음. 교수만 가지고 있으니 따로 만들어줌
    def talk(self):
        print(f'반갑네. 나는{self.name}이라고 하네. 담당은 {self.department}라네.')


    p1 = Professor('김교수', 52, '컴퓨터공학과')
    p1.talk()
    p1.eat
```

```python
class A:
    v = 'A'

class B(A):
    v = 'B'

class C(A):
    v = 'C'

class D(B,C):
    pass

d = D()
print(d.v)
```
