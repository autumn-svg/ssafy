if
조건문의 기본 형태, 조건 만족 시 내부 코드 실행하고 표현식으로 작성

elif
이전의 조건을 만족하지 못 하고 추가로 다른 조건이 필요할 때 사용, 여러 개 사용 가능

else
모든 조건들을 만족하지 않으면 실행됨

```python
# 반복문 for
# 반복 가능한 객체의 요소들을 반복하는데 주로 사용

#반복 가능한 객체란 요소를 하나씩 반환할 수 있는 모든 객체, 
# 시퀀스(순서가 있는) 뿐만 아니라 비시퀀스(dict, set)등도 반복 가능
# 숫자형(int, float) 은 정수형 range()로 반복

#iterable 요소를 하나씩 순회하며 반복, 횟수가 명확하게 정해져 있으면 유용


    for 변수 in 반복 가능한 객체: 
        코드 블록

student_list = ['Alice', 'Bob', 'Charlie'] #재료가 3개라서 3번 반복됨

for student in student_list:
    print(f"Hello, {student}!") 
```

```python
#예시 문제
#첫 항목이 반복변수item에 할당되고 코드 블록이 실행
#다음 반복변수에 다음 항목이 할당...

#for 다음에 작성하는 임시변수는 단수형으로 사용하는 것을 권장

item_list = ['apple', 'banana', 'coconut']

for item in item_list: #item: 반복변수
    print(item)

#출력
'''
apple
banana
coconut
'''
```

```python
#리스트에서 첫번째 짝수만 찾고 반복 종료하기

numbers = [1, 3, 5, 6, 7, 9, 10, 11]
found_even = False #found_even은 깃발 변수라고도 함, 상태 확인이라고 생각하면 됨

for num in numbers:
    if num % 2 ==0:
        print('첫번째 짝수를 찾았습니다:', num)
        found_even = True
        break

if not found_even:
    print('짝수를 찾지 못했습니다.')
```

```python
def check_sign(num):
    if num < 0:
        return '음수'
    elif num > 0:
        return '양수'
    elif num == 0:
        return 0
   


print(check_sign(5)) # 양수
print(check_sign(-10)) # 음수
print(check_sign(0)) # 0

```

```python
def print_odd(lst):
    for i in lst:
        if i % 2 == 1: 
            print(i)

print_odd([1,2,3,4,5,99]) 
# 1
# 3
# 5
# 99
```

```python
def is_even_or_odd(num):
    if num % 2 == 1:
        return '홀수'
    else :
        return '짝수'

    
print(is_even_or_odd(2)) # 짝수
print(is_even_or_odd(5)) # 홀수
print(is_even_or_odd(0)) # 짝수
```

```python
#딕셔너리 순회
#딕셔너리는 키를 추출해 낸다

my_dict = {
    'x': 10,
    'y': 20,
    'z': 30,
}

for key in my_dict: #for value라고 써도 value안에 키 값이 들어 감
    print(key)
    print(my_dict[key])
```

```python
#인덱스로 리스트 순회

#인덱스로 접근해서 해당 요소를 변경하기

numbers = [4, 6, 10, -8, 5] # index 0 1 2 3 4

for i in range(len(numbers)): #i로 한 이유: 그냥 인덱스라고 표시하려고
    numbers[i] = numbers[i] * 2

print(numbers)   #[8, 12, 20, -16, 10]
```

```python
#중첩된 반복문

outers = ['A', 'B']
inners = ['c', 'd']

for outer in outers:
    for inner in inners:
        print(outer, inner) #a c, a d, b c, b d

#출력은?
#총 네 바퀴 도는 코드
```

```python
map(function, iterable) #map, function함수, iterable 인자

numbers = [1, 2, 3]
result = map(str, numbers) #str인자에 numbers가 들어 감. 그래서 문자열로 바꾸겟다 이거임. like 반복문 같음

print(result) #<map object at 0x00000239C915D760> 덩어리로 나오기 때문에
print(list(result)) #['1', '2', '3'] 리스트로 내보내야 함
```
