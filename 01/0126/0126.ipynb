```python
#2997. 문자열과 리스트 메서드 3Lv3

def restructure_word(word, arr):
    for c in word:
        if c.isdecimal():
            for i in range(int(c)):
                arr.pop()
        else :
            arr.remove(c)
        return arr

original_word = '코딩 공부는ㄴ 1일ㄹ 1커ㅓ밋ㅅ @@@#^()#_+!&~:"'
word = '1ㄴ2ㄹ3ㅓ4ㅅ5'
arr = []

arr.extend(original_word)

result = restructure_word(word, arr)
 #문자열 하나하나 리스트에 넣으려면 extend
print (arr)
print("".join(result))
```
